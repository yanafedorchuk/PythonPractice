```python
import numpy as np
import pandas as pd
from numpy import nan as NA
```

```python
airport = pd.read_csv('airports.csv', names = ['Aiport', 'Name', 'City', 'Country', 'IATA', 'ICAO', 'Latitude', 'Longitude', 'Altitude', 'Timezone', 'DST', 'Tz database time zone', 'Type', 'Source'])
airport
```

```python
AUS=['Australia']
```

```python
new=airport[airport['Country'].isin(AUS)]
new
```

```python
new=new.replace('\\N',NA)
new
```

```python
air_aus = new[new['IATA'].str.len()==3]
air_aus
```

```python
route = pd.read_csv('routes.csv',names = ['Airline', 'AirlineID', 'SourceAirport', 'SourceAirportID', 'DestinationAirport', 'DestinationAirportID', 'Codeshare', 'Stops', 'Equipment'])
route1=route.replace('\\N',NA)
```

```python
route_aus=route1[(route1['SourceAirport'].isin(air_aus['IATA']))&(-route1['DestinationAirport'].isin(air_aus['IATA']))]
route_aus
```

```python
source_airport = pd.DataFrame(route_aus.groupby('SourceAirport').size())
destination_airport = pd.DataFrame(route_aus.groupby('DestinationAirport').size())

to_merge=pd.DataFrame(airport, columns=['IATA','Country'])
to_merge=to_merge.rename(columns = {'IATA': 'DestinationAirport'})
merged=route_aus.merge(to_merge, on='DestinationAirport', how='left')
destination_airport
```
