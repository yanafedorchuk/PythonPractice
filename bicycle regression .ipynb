```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#using data from https://archive.ics.uci.edu/ml/machine-learning-databases/00275/
bicycle=pd.read_csv('day.csv')
bicycle
```

```python
sns.regplot(data=bicycle, x='temp', y='cnt')
```

```python
sns.regplot(data=bicycle, x='hum', y='cnt')
```

```python
sns.regplot(data=bicycle, x='windspeed', y='cnt')
```

```python
sns.residplot(data=bicycle, x='temp', y='cnt')
```

```python
sns.regplot(data=bicycle, x='temp', y='cnt', order=2)
```

```python
sns.residplot(data=bicycle, x='temp', y='cnt', order=2)
```

```python
sns.regplot(data=bicycle, x='mnth', y='cnt', x_jitter=.1, order=2)
```

```python
sns.regplot(data=bicycle, x='mnth', y='cnt', x_estimator=np.mean, order=2)
```

```python
sns.regplot(data=bicycle, x='temp', y='cnt', x_bins=4)
```

```python
pd.crosstab(bicycle['mnth'], bicycle['weekday'], values=bicycle['cnt'],
            aggfunc='mean').round(0)
```

```python
sns.heatmap(pd.crosstab(bicycle['mnth'], bicycle['weekday'], values=bicycle['cnt'],
            aggfunc='mean'))
```

```python
sns.heatmap(pd.crosstab(bicycle['mnth'], bicycle['weekday'], values=bicycle['cnt'],
            aggfunc='mean'), annot=True, cmap='YlGnBu',
           cbar=False, linewidths=.5)
```

```python
sns.heatmap(bicycle.corr())
```
