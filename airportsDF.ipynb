```python
import numpy as np
import pandas as pd
from numpy import nan as NA
```

```python
#отобрать только европейские аэропорты
airport = pd.read_csv('airports.csv', names = ['Aiport', 'Name', 'City', 'Country', 'IATA', 'ICAO', 'Latitude', 'Longitude', 'Altitude', 'Timezone', 'DST', 'Tz database time zone', 'Type', 'Source'])
airport.set_index('IATA')
del airport['Aiport']
```

```python
Europe=np.array(['Belgium', 'Germany', 'Estonia', 'Finland', 'United Kingdom',
'Guernsey', 'Jersey', 'Isle of Man', 'Netherlands', 'Ireland',
'Denmark', 'Luxembourg', 'Norway', 'Poland', 'Sweden', 'Spain',
'Albania', 'Bulgaria', 'Cyprus', 'Croatia', 'France', 'Greece',
'Hungary', 'Italy', 'Slovenia', 'Czech Republic', 'Malta',
'Austria', 'Portugal', 'Bosnia and Herzegovina', 'Romania',
'Switzerland', 'Turkey', 'Moldova', 'Macedonia', 'Gibraltar',
'Serbia', 'Montenegro', 'Slovakia', 'Ukraine', 'Russia', 'Belarus',
'Latvia', 'Lithuania','Faroe Islands', 'Greenland', 'Iceland'])
```

```python
new=airport[airport['Country'].isin(Europe)]
new
```

```python
new1=new.replace('\\N',NA)
new1
```

```python
air_eu = new1[new1['IATA'].str.len()==3]
air_eu
```

```python
route = pd.read_csv('routes.csv',names = ['Airline', 'AirlineID', 'SourceAirport', 'SourceAirportID', 'DestinationAirport', 'DestinationAirportID', 'Codeshare', 'Stops', 'Equipment'])
route1=route.replace('\\N',NA)
```

```python
route_eu=route1[(route1['SourceAirport'].isin(air_eu['IATA']))|(route1['DestinationAirport'].isin(air_eu['IATA']))]
route_eu
```

```python
source_airport = pd.DataFrame(route_eu.groupby('SourceAirport').size())
destination_airport = pd.DataFrame(route_eu.groupby('DestinationAirport').size())
merged = pd.merge(source_airport, destination_airport, left_index=True, right_index=True)
merged
```

```python
merged['Fly'] = merged['0_x'] + merged['0_y']
del merged['0_x']
```

```python
del merged['0_y']
```

```python
merged.index
```

```python
air_eu=airport.set_index('IATA')
```

```python
pivot_table= pd.DataFrame(air_eu, columns = ['Country', 'Name', 'Fly'])
pivot_table = pivot_table.sort_values(by=['Country'], ascending=True)
pivot=pd.merge(pivot_table, merged, left_index=True, right_index=True)
del pivot['Fly_x']
pivot
```

```python
pivot = pivot.rename(columns = {'Fly_y': 'Fly'})
pivot
```

```python
pivot1=pivot[pivot['Country'].isin(Europe)]
pivot1
```
