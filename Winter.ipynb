```python
winter1 = pd.read_csv('winter.csv')
winter1.head()
```

```python
winter = winter1[winter1['Year']==1976]
```

```python
men = np.array(winter[winter['Gender']=='Men']['Discipline'].unique())
men
```

```python
women = np.array(winter[winter['Gender']=='Women']['Discipline'].unique())
women
```

```python
#Дисциплины, в которых разыгравались только мужские комплекты наград
np.setdiff1d(men,women)
```

```python
#Общее кол-во медалей за все годы мужчинами и женщинами
men = winter1[winter1['Gender']=='Men'].groupby('Medal').count()
men
```

```python
women = winter1[winter1['Gender']=='Women'].groupby('Medal').count()
women
```

```python
medals_women=pd.DataFrame(women['Year'])
medals_women=medals_women.rename(columns={'Year': 'Women'})
medals_women
```

```python
medals_men=pd.DataFrame(men['Year'])
medals_men=medals_men.rename(columns={'Year': 'Men'})
medals_men
```

```python
all_medals = pd.concat([medals_men, medals_women], axis=1, join='inner')
all_medals
```
