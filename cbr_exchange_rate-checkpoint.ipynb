```python
import numpy as np
import pandas as pd
```

```python
html='https://www.cbr.ru/currency_base/dynamics/?UniDbQuery.Posted=True&UniDbQuery.mode=1&UniDbQuery.date_req1=&UniDbQuery.date_req2=&UniDbQuery.VAL_NM_RQ=R01235&UniDbQuery.FromDate=08.12.2016&UniDbQuery.ToDate=15.12.2018'
df=pd.read_html(html, header=1)[0]
df.head()
df=df.rename(columns={'Дата': 'Date', 'Единиц':'Units', 'Курс':'Exchange_rate'})
df['Exchange_rate']=df['Exchange_rate']/10000
df['Date']=pd.to_datetime(df['Date'])
df=df.set_index('Date')
df.head(5)
```

```python
range_fri=pd.date_range(start='2016-08-12', end='2018-12-15', freq='W-FRI')
```
